```python
import numpy as np
```

```python
m = pd.read_csv('sonos_stock_8_30.csv')
m = m[(m.interval == 30)]
```

```python
art_df = pd.read_csv('apple_articles_20180822_20180829_df.csv')
```

```python
art_df = art_df.drop('Unnamed: 0', axis =1)
```

```python
art_df.index = art_df.publishedAt
art_df.index = pd.to_datetime(art_df.index)
art_df = art_df.drop('publishedAt',axis=1)
```

```python
art_df['day'] = art_df.index.date
```

```python
new = pd.DataFrame(index=m['Unnamed: 0'])
new.index = pd.to_datetime(new.index)
new['day'] = new.index.date
```

```python
art_df.source = art_df.source.fillna('Other')
```

```python
new.head(1)
```

```python
values = []
for index, row in new.iterrows():
    avg = art_df[(art_df.index <= index)].groupby('day', as_index=False)['compound'].mean()
    avg.day = pd.to_datetime(avg.day)
    value = avg.loc[avg['day'] == str(row.day)]['compound'].values
    if len(value) > 0:
        values.append(value[0])
    else:
        values.append(np.mean(avg['compound'].values))

new['day compound avg'] = values
```

```python
values = []
for index, row in new.iterrows():
    holder = []
    avg = pd.DataFrame(art_df[(art_df.index <= index)]['neu'])
    avg.index = pd.to_datetime(avg.index)
    avg['day'] = avg.index.date
    avg.day = pd.to_datetime(avg.day)
    day_of = avg[(avg['day'] == str(row.day))]['neu'].values
    day_before = avg.loc[avg.day == str((row.day - timedelta(days=1)))]['neu'].values
    holder.extend(day_of)
    holder.extend(day_before)
    if len(holder) > 0:
        values.append(np.mean(holder))
    else:
        values.append(np.mean(avg['neu'].values))

new['two day neutral avg'] = values
```
